Train test split


```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(
    df.drop('HeartDisease',axis=1),
    df['HeartDisease'],
    test_size=0.2,
    random_state=42,
    stratify=df['HeartDisease']
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import numpy as np

# List of solvers to test
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']

# To store test scores
test_score = np.zeros(len(solver))

best_solver = None

# Try each solver
for i, n in enumerate(solver):
    lr = LogisticRegression(solver=n, max_iter=1000)
    lr.fit(X_train, y_train)

    test_score[i] = lr.score(X_test, y_test)

    if test_score[i] == test_score.max():
        best_solver = n

lr = LogisticRegression(solver=best_solver, max_iter=1000)
lr.fit(X_train, y_train)

lr_pred = lr.predict(X_test)

print("Best Solver:", best_solver)
print("LogisticRegression Score:", accuracy_score(y_test, lr_pred))
```

```python
import pickle

with open('LogisticR.pkl', 'wb') as f:
    pickle.dump(lr, f)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import f1_score

kernels={'linear':0,'poly':0,'rbf':0,'sigmoid':0}
best=''
for i in kernels:
  svm=SVC(kernel=i)
  svm.fit(X_train,y_train)
  pred=svm.predict(X_test)
  kernels[i]=f1_score(y_test,pred,average="weighted")
  if kernels[i]==max(kernels.values()):
    best=i
svm=SVC(kernel=best)
svm.fit(X_train,y_train)
svm_pred=svm.predict(X_test)
print(f'SVM f1_score kernel({best}): {f1_score(y_test,svm_pred,average="weighted")}')
```

```python
with open('SVMR.pkl', 'wb') as f:
    pickle.dump(svm, f)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Create Decision Tree model
dtree = DecisionTreeClassifier(class_weight='balanced', random_state=42)

# Hyperparameter grid
param_grid = {
    'max_depth': [3, 4, 5, 6, 7, 8],
    'min_samples_split': [2, 3, 4],
    'min_samples_leaf': [1, 2, 3, 4]
}

# Grid Search
grid_search = GridSearchCV(
    dtree,
    param_grid,
    cv=5,
    scoring='accuracy'
)

# Train
grid_search.fit(X_train, y_train)

# Train best model
dtree_best = DecisionTreeClassifier(
    **grid_search.best_params_,
    class_weight='balanced',
    random_state=42
)

dtree_best.fit(X_train, y_train)

# Predictions
dt_pred = dtree_best.predict(X_test)

# Accuracy
print("Decision Tree Accuracy:", accuracy_score(y_test, dt_pred))
```

```python
with open('DecisionR.pkl', 'wb') as f:
    pickle.dump(dtree_best, f)
```
